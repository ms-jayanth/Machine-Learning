### 2. Using Euclideon distance as a metric for UBCF 

```python
from sklearn.metrics import pairwise_distances
```

```python
euclideon_recommendation=pd.pivot_table(data=user_data,values='rating',index='userId',columns='movie').fillna(0)
euclideon_recommendation
```

```python
euclidean_metric=pd.DataFrame(pairwise_distances(X=euclideon_recommendation, metric='euclidean').round(2))
euclidean_metric.columns=user_data['userId'].unique()
euclidean_metric.index=user_data['userId'].unique()
euclidean_metric
```

### 3. Using Cosine  distence as a metric for UBCF 

```python
cosine_recommendation=euclideon_recommendation.copy()
cosine_recommendation
```

```python
cosine_metric=pd.DataFrame(1-pairwise_distances(X=cosine_recommendation, metric='cosine').round(2))
cosine_metric.columns=user_data['userId'].unique()
cosine_metric.index=user_data['userId'].unique()
cosine_metric
```
