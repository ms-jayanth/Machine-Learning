## 4. Data Preparation

#### We are choosing only United Kingdom data

```python
uk_data=online_retail_data[online_retail_data['Country']=='United Kingdom']
uk_data
```

```python
uk_data.isna().sum()
```

Since Customer id is not required for our model building we will drop that column and Description's null value percentage is 1454/495478=0.0029 will not effect our model building if we drop null value observations(Description Col'n is very important)

```python
del uk_data['CustomerID']
```

```python
uk_data.dropna(axis=0, inplace=True)
```

```python
uk_data.shape
```

```python
uk_data.isna().sum()
```

```python
#Data Preparation for Apriori
uk_data_pivot_table = pd.pivot_table(data = uk_data,index='InvoiceNo',columns='Description', values='Quantity').fillna(0)
uk_data_pivot_table
```

```python
#One Hot Data using applymap
uk_encoded=uk_data_pivot_table.applymap(lambda x: 1 if x>=1 else 0)
uk_encoded
```

## 5. Model Building

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
freq_item_set=apriori(df=uk_encoded,min_support=0.02, use_colnames=True )
freq_item_set
```

```python
best_associates=association_rules(df=freq_item_set,metric='confidence',min_threshold=0.5)
best_associates
```

```python
best_associates.to_csv('Best_Associates.csv')
```
