## 1. Import Necessary Libraries

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

## 2. Import Data

```python
claimants_data=pd.read_csv('claimants.csv')
claimants_data.head(40)
```

## 3. Data Understanding

```python
claimants_data.shape
```

```python
claimants_data.isna().sum()
```

```python
claimants_data.dtypes
```

```python
claimants_data.describe()
```

## 4. Data Preparation

```python
del claimants_data['CASENUM']
```

```python
claimants_data.head()
```

```python
claimants_data.dropna(axis=0, inplace=True)
```

```python
claimants_data.shape
```

```python
claimants_data.isna().sum()
```

```python
claimants_data.dtypes
```

## 5. Model Building

```python
X=claimants_data.drop(labels='ATTORNEY', axis=1)
y=claimants_data[['ATTORNEY']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.20, random_state=12, shuffle=True)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

## 6. Model Training

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_model=DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
```

```python
y_pred_train=decision_tree_model.predict(X_train)
y_pred_train
```

### Plot the Tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(15,10))
plot_tree(decision_tree= decision_tree_model, filled=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
print('Accuracy Score:',accuracy_score(y_train, y_pred_train))
```

```python
print('Classification Report:\n',classification_report(y_train, y_pred_train))
```

```python
print('Confusion Matrix:\n',confusion_matrix(y_train, y_pred_train))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr, tpr, thresholds = roc_curve(y, decision_tree_model.predict_proba (X)[:,1])

auc = roc_auc_score(y_train,y_pred_train)
print(auc)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

## 7. Test Data

```python
y_pred_test=decision_tree_model.predict(X_test)
y_pred_test
```

```python
print('Accuracy Score:',accuracy_score(y_test, y_pred_test))
```

```python
print('Classification Report:\n',classification_report(y_test, y_pred_test))
```

```python
print('Confusion Matrix:\n',confusion_matrix(y_test, y_pred_test))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr, tpr, thresholds = roc_curve(y_test, decision_tree_model.predict_proba (X_test)[:,1])

auc = roc_auc_score(y_test,y_pred_test)
print(auc)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

## ------------------------------------------------------------------------------------------------------------------------------

## BIAS - VARIANCE

* Training Error/Accuracy - Bias
* Test Error/Accuracy - Variance.

### Model Overfitting - Less Bias and High Variance.
### Model Underfitting - High Bias and Less Variance.

### EXPECTED MODEL -- GENERALIZED MODEL - Less Bias and Less Variance.

So, there is always a **Tradeoff maintaned between Bias and Variance.**

## ------------------------------------------------------------------------------------------------------------------------------

### We are facing Overfitting Problem

## GridSearchCV Technique:

```python
#Exhaustive search over specified parameter values for an estimator.
from sklearn.model_selection import GridSearchCV

grid_search=GridSearchCV(estimator=decision_tree_model, 
                         param_grid={'criterion':['entropy','gini'],
                                     'max_depth':[2,3,4,5,6,7,8,9,10]}, 
                         cv=5)

grid_search.fit(X,y)
print(grid_search.best_params_)
print(grid_search.best_score_)
```
